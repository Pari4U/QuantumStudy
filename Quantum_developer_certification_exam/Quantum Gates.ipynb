## Single qubit Gates

```python
import numpy as np
import math
from math import pi, sqrt
# Importing standard Qiskit libraries
from qiskit import *
#from qiskit.tools.jupyter import *
from qiskit.visualization import *
from qiskit.quantum_info import random_statevector
from qiskit.quantum_info import Statevector
```

### Pauli X, Y and Z-gate

#### X-gate

```python
qc = QuantumCircuit(1)
qc.x(0) 
qc.draw(output='mpl')
```

```python
sv = Statevector.from_int(0,2**1) 
sv_ev = sv.evolve(qc)
sv_ev.draw('latex')
```

```python
qc.measure_all()
qasm_sim = BasicAer.get_backend('qasm_simulator')
result = execute(qc, qasm_sim).result()
counts = result.get_counts() 
plot_histogram(counts) 
```

```python
back = BasicAer.get_backend('statevector_simulator')
result = execute(qc, back).result()
sv = result.get_statevector()
plot_bloch_multivector(sv)
```

### GHZ circuit

```python
ghz_circuit = QuantumCircuit(3)
ghz_circuit.h(0)
ghz_circuit.cx([0,0],[1,2]) 

ghz_circuit.draw(output='mpl')
```

```python
sv = Statevector.from_int(0,2**3) 
sv_ev = sv.evolve(ghz_circuit)
sv_ev.draw('latex')
```

### Qiskit hacks 101

```python
qc1 = QuantumCircuit(2)
qc1.h(0)
qc1.x(0)
qc1.s(1)
qc1.y(1)
qc1.draw(output='mpl')
```

```python
qc2 = QuantumCircuit(2)
qc2.h(0)
qc2.cx(0,1)
qc2.draw(output='mpl')
```

#### if we have two circuits of same qubits, for appending them in a single circuit we need to simply use  '+' operator 

```python
new_qc1 = qc1+qc2
```

```python
new_qc1.draw(output='mpl')
```

```python
# the depth of the circuit is given by 
qc1.depth()
```

#### if we have two circuits of different qubits, for appending them we can use compose () function in qiskit

```python
new_qc2 = ghz_circuit.compose(qc2,[0,1])
new_qc2.draw(output='mpl')
```

#### QASM to Quantum CIrcuit and vice-versa

```python
# To convert quantum circuit to qasm 
new_qc2.qasm(formatted='True')
```

```python
# converting quantum circuit to qasm and saving it to file
new_qc2.qasm(formatted='True',filename='new_qc2.qasm')
```

```python
# converting qasm file to quantum circuit 
new_qc = QuantumCircuit.from_qasm_file('new_qc2.qasm')
new_qc.draw(output='mpl')
```
